# Look at trained models

```python
from ligo_emu import mlp
from ligo_emu import data_module

# ckpt_file_path = "../model_outputs/mlp_cosmo_to_pk_dims_128_128_128/epoch=99-val_loss=386410.41.ckpt"
# ckpt_file_path = "../model_outputs/mlp_cosmo_to_pk_dims_128_128_128/epoch=99-val_loss=908205.75.ckpt"

# ckpt_file_path = "../model_outputs/mlp_cosmo_to_pk_dims_256_256_256_256/epoch=99-val_loss=387392.90.ckpt"
# ckpt_file_path = "../model_outputs/mlp_cosmo_to_pk_dims_256_256_256_256/epoch=99-val_loss=-7702.86.ckpt"

ckpt_file_path = "../model_outputs/model_256_256/epoch=99-val_loss=0.03.ckpt"

model = mlp.MLP.load_from_checkpoint(ckpt_file_path)

dataset = data_module.LIGODataset(
    model.hparams['val_path'],
    model.hparams['norm_path'],
    None,
)
```

```python
import torch 

data_path = '../data/dataset.h5'
# data_path = '../data/powerspectra_sigma8_z-f-b1_cA_val.h5'

with h5py.File(data_path, 'r') as hf:
    print(hf.keys())
    x = hf['x'][:]
    y = hf['y'][:].flatten()


y[y < dataset.y_min] = dataset.y_min

x = dataset.scale_x(x)
# y = dataset.scale_y(y)

y_pred = model(torch.tensor(x)).detach().numpy().flatten()
y_pred = dataset.scale_y(y_pred, inv=True)
```

```python
ymin = min(np.concatenate([y, y_pred]))
ymax = max(np.concatenate([y, y_pred]))

bins = np.linspace(ymin-1, ymax+1, 50)

plt.hist2d(y, y_pred, cmap=plt.cm.Blues, bins=bins);
plt.xlabel('y')
plt.ylabel('y_pred')
plt.colorbar()
```
